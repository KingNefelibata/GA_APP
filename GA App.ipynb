## Question 2:
## Create a histogram that shows the distribution for number of backers. What is the skew of the distribution?

```python
mean = df['backers'].mean()
std = df['backers'].std(ddof=1)
median = df['backers'].median() 
mode = df['backers'].mode()
print("the mean is {}".format(mean))
print("the mode is {}".format(mode))
print("the std is {}".format(std))
print("the median is {}".format(median))
```

```python
# The median is less than the mean so it may infer a Positive Skew 
```
